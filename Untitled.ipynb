```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a=np.arange(1,1000)
b=np.arange(1,1000)

y_ab=a*b
y_a_b=a*np.log2(a)+b*np.log2(b)+a*np.log2(b)
fig = plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(a,y_ab,label='a*b')
ax.plot(a,y_a_b,label='a_b')
ax.legend()
```

```python
from bs4 import BeautifulSoup as Soup
import pandas as pd
import hdbscan
map_kml="doc.kml"
with open(map_kml) as data:
    kml_soup=Soup(data,'lxml-xml')
coords=kml_soup.find_all('coordinates')
coords_list=[]
for x in coords:
    coords_list.append(str(x))
names=kml_soup.find_all('name')
names_list=[]
for x in names:
    names_list.append(str(x)) 
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\s+', gaps=True)
treated=[]
for p in coords_list:
    treated.append(tokenizer.tokenize(p))
    treated_2=[]
for i in range (0,len(treated)):
    treated_2.append(treated[i][1])
tokenizer = RegexpTokenizer(',', gaps=True)
treated_3=np.zeros((29,2))
for i in range(0,len(treated_2)):
    treated_3[i][0]=(tokenizer.tokenize(treated_2[i]))[0]
    treated_3[i][1]=(tokenizer.tokenize(treated_2[i]))[1]
tokenizer = RegexpTokenizer('>|<', gaps=True)
treated=[]
for p in names_list:
    treated.append(tokenizer.tokenize(p))
treated_2=[]
for i in range (0,len(treated)):
    treated_2.append(treated[i][1])
d = {'Longitude': treated_3[:,1], 'Latitude': treated_3[:,0],'Landmark':treated_2}
df = pd.DataFrame(data=d)
df_ct = pd.DataFrame(data=d)
```

```python
df['Latitude']=(df['Latitude']-df['Latitude'].mean())/(df['Latitude'].max()-df['Latitude'].min())
df['Longitude']=(df['Longitude']-df['Longitude'].mean())/(df['Longitude'].max()-df['Longitude'].min())
```

```python
from sklearn.cluster import KMeans 
kmeans_1 = KMeans(n_clusters=4,n_jobs=-1)
X = df[['Longitude','Latitude']].values
predictions = kmeans_1.fit_predict(X) 
fig = plt.figure(figsize=(25,25))
ax = fig.add_subplot(111)
scatter = ax.scatter(df_ct['Latitude'],df_ct['Longitude'],
                     c=predictions,s=100)
ax.set_title('K-Means Clustering')
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitutde')
plt.colorbar(scatter)
df_ct['Cluster']=predictions
for i in range(0,29):
    ax.annotate(df_ct['Landmark'][i], (df_ct['Latitude'][i],df_ct['Longitude'][i]))
fig.savefig('k-means.png')
df_ct.sort_values(by=['Cluster'])

```

```python
rads = np.radians(X)
clusterer = hdbscan.HDBSCAN(min_cluster_size=2, metric='haversine',)
predictions = clusterer.fit_predict(rads)
fig = plt.figure(figsize=(25,25))
ax = fig.add_subplot(111)
scatter = ax.scatter(df_ct['Latitude'],df_ct['Longitude'],
                     c=predictions,s=100)
ax.set_title('HDBSCAN Clustering')
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitutde')
plt.colorbar(scatter)
df_ct['Cluster']=predictions
for i in range(0,29):
    ax.annotate(df_ct['Landmark'][i], (df_ct['Latitude'][i],df_ct['Longitude'][i]))
fig.savefig('HDSBSCAN.png')
df_ct.sort_values(by=['Cluster'])
```

```python
from sklearn.cluster import DBSCAN 
kmeans_1 = DBSCAN(eps=0.1,min_samples=2,n_jobs=-1)
X = df[['Longitude','Latitude']].values
predictions = kmeans_1.fit_predict(X) 
fig = plt.figure(figsize=(25,25))
ax = fig.add_subplot(111)
scatter = ax.scatter(df_ct['Latitude'],df_ct['Longitude'],
                     c=predictions,s=100)
ax.set_title('Sklearn DBSCAN Clustering')
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitutde')
plt.colorbar(scatter)
df_ct['Cluster']=predictions
for i in range(0,29):
    ax.annotate(df_ct['Landmark'][i], (df_ct['Latitude'][i],df_ct['Longitude'][i]))
fig.savefig('SKLEARN-DBSCAN.png')
df_ct.sort_values(by=['Cluster'])
```
